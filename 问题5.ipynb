要比较电动车的耗电量和燃油车的油耗，我们可以使用等效能源单位，通常是千瓦时（kWh）和升或加仑。以下是一个简单的换算方法：

1. **电动车的耗电量：** 电动车的能耗通常以千瓦时每百公里（kWh/100 km）或者英里（kWh/100 miles）来表示。

2. **燃油车的油耗：** 燃油车的油耗通常以升每百公里（L/100 km）或者英里（miles per gallon, mpg）来表示。

为了进行比较，我们可以使用以下换算关系：

- 1千瓦时等于3.6兆焦耳（Mj）。
- 1升汽油的能量大约等于31.536千瓦时。

然后，我们可以使用以下公式来比较两者的相对能效：

$ \text{相对能效} = \frac{\text{电动车的耗电量}}{\text{燃油车的油耗} \times \text{能量换算系数}} $

这个相对能效的值越大，表示电动车在单位距离上的能效越高，即更为节能。需要注意的是，这只是一个简单的换算方法，实际上还有其他因素需要考虑，例如能源生产过程中的损耗、电池制造和处理的环境影响等。

请注意，上述数值和换算系数是一般的估算，实际的能效和能源转换效率可能因车型、驾驶条件和能源来源而有所不同。

### 燃油 知识点

燃油的密度通常用于将体积单位（升，L）转换为质量单位（千克，kg）。密度是指单位体积内的质量，对于燃油而言，单位可以是千克每升（kg/L）。

给定燃油的密度为 \(0.725 \, \text{kg/L}\)，我们可以使用这个密度进行换算。

1. **从升到千克：**
   $ \text{燃油质量（kg）} = \text{燃油体积（升）} \times \text{燃油密度（kg/L）} $

2. **从千升（kl）到千克：**
   $ \text{燃油质量（kg）} = \text{燃油体积（千升）} \times \text{燃油密度（kg/L）}$

这两个公式可以帮助你在升和千升之间进行换算。

请注意，这里的燃油密度 $0.725 \, \text{kg/L}$ 是一个假设值，实际的燃油密度可能会在不同的燃油种类之间有所变化。

### 给出新能源车和油耗车，试问具体能耗减少过程运算

要计算相对节能量，首先需要将电动车和燃油车的能耗转换为标准煤（Standard Coal Equivalent, SCE）。折标准煤系数表示单位质量的燃油或电能所相当于的标准煤量。

给定的数据为：

- 车用成品油密度 \(0.725 \, \text{kg/L}\)
- 燃油折标准煤系数 \(1.4714 \, \text{kgce/kg}\)
- 电能折标准煤系数 \(0.1229 \, \text{kgce/kWh}\)
- 电动车的耗电量 \(14,300 \, \text{Wh/100km}\)
- 燃油车的油耗 \(39,000 \, \text{L/100km}\)

**步骤 1：将耗电量和油耗转换为标准煤。**

1. 电动车的标准煤耗电量（以千瓦时为单位）：
   $ \text{电动车标准煤耗电量} = \text{电动车的耗电量} \times \text{电能折标准煤系数} $
   $ \text{电动车标准煤耗电量} = 14,300 \, \text{Wh/100km} \times 0.1229 \, \text{kgce/kWh} $

2. 燃油车的标准煤油耗（以千升为单位）：
   $ \text{燃油车标准煤油耗} = \text{燃油车的油耗} \times \text{车用成品油密度} \times \text{燃油折标准煤系数} $
   $ \text{燃油车标准煤油耗} = 39,000 \, \text{L/100km} \times 0.725 \, \text{kg/L} \times 1.4714 \, \text{kgce/kg} $

**步骤 2：计算相对节能量。**

$ \text{相对节能量/t标准煤} = \frac{\text{燃油车标准煤油耗}}{\text{电动车标准煤耗电量}} $

将上述计算过程中的数值代入这个公式，即可得到相对节能量的值。

```python
cost = pd.DataFrame({'车辆类型':['公交车','出租车','物流车','旅游车','环卫车','私家车'],'电动车型耗电量/万kWh':[1.1,3.0,5.0,8.8,2,0.4],'对应燃油车型油耗/kL':[30,15,20,24,10,2]})
cost
```

通过计算，我们可以算出这100万城市所有车辆类型的电动车型耗电量和燃油车型油耗量，进而算出相对节能量标准煤
